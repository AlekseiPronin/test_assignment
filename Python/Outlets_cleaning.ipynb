```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine
from config import host, user, password, db_name
import numpy as np
```

```python
# DB query

engine = create_engine(f'mysql+pymysql://{user}:{password}@{host}:3307/{db_name}')    

#read the entire table
df = pd.read_sql('SELECT * FROM outlets_copy', engine) 
df
```

```python
df['Торг_точка_грязная_адрес'] = df['Торг_точка_грязная_адрес'].str.replace(r'[^\w\s]+', ' ')
df.head(20)
```

```python
df['Торг_точка_грязная'] = df['Торг_точка_грязная'].str.replace(r'[^\w\s]+', ' ')
df.head(20)
```

```python
# removing repeated values

df = df[df.Торг_точка_грязная_адрес != 'он же']
df
```

```python
# dropping duplicates
with_empty_address = df.drop_duplicates()
with_empty_address
```

```python
# assigning NaNs to rows with no addresses
df['Торг_точка_грязная_адрес'].replace(' ', np.nan, inplace=True)
df
```

```python
# dropping NaNs from Торг_точка_грязная_адрес
df.dropna(subset=['Торг_точка_грязная_адрес'], inplace=True)
df
```

```python
# assigning new id to outlet_clean_id based on df with no empty addresses
df = df.reset_index()
df['outlet_clean_id'] = df.index + 1
df.drop('index', axis=1)
```

```python
# creating outlets_clean df to fit sql table
outlets_clean = df[['outlet_clean_id', 'Торг_точка_грязная_адрес']]

outlets_clean = outlets_clean.rename(columns={'outlet_clean_id': 'id', 
                                              'Торг_точка_грязная_адрес': 'Торг_точка_чистый_адрес'})
outlets_clean
```

```python
with_empty_address['outlet_clean_id'] = with_empty_address['id']
with_empty_address.head()

outlets_clean_no_address = with_empty_address[['outlet_clean_id', 'Торг_точка_грязная_адрес']]

outlets_clean_no_address = outlets_clean_no_address.rename(columns={'outlet_clean_id': 'id', 
                                              'Торг_точка_грязная_адрес': 'Торг_точка_чистый_адрес'})
outlets_clean_no_address
```

```python
# saving data to sql table 'outlets_clean'
outlets_clean.to_sql('outlets_clean', engine, if_exists='append', index=False)
```

```python
# saving data to sql table 'outlets_clean_no_address'
outlets_clean_no_address.to_sql('outlets_clean_no_address', engine, if_exists='append', index=False)
```
